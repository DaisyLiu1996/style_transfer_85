```python
import matplotlib.pyplot as plt
import re
```

```python
filename = '/datasets/home/79/979/hay004/ECE285/little285/checkpoints/loss_log.txt'
X= []
with open(filename, 'r') as f:
    lines = f.readlines()
    for line in lines:
        X.append(line)
X = X[1:]
print(X[2:5])
```

```python
D_A = []
G_A = []
cycle_A = []
idt_A = []
D_B = []
G_B = []
cycle_B = []
idt_B = []
for x in X:
    loss = re.match('.*D_A: (.*) G_A: (.*) cycle_A: (.*) idt_A: (.*) D_B: (.*) G_B: (.*) cycle_B: (.*) idt_B: (.*) ', x)
    D_A.append(float(loss.group(1)))
    G_A.append(float(loss.group(2)))
    cycle_A.append(float(loss.group(3)))
    idt_A.append(float(loss.group(4)))
    D_B.append(float(loss.group(5)))
    G_B.append(float(loss.group(6)))
    cycle_B.append(float(loss.group(7)))
    idt_B.append(float(loss.group(8)))
```

```python
fig, axs = plt.subplots(2,4,figsize=(10,8),sharex=True)
n = len(D_A)
axs[0][0].plot(list(range(n)),D_A)
axs[0][1].plot(list(range(n)),G_A)
axs[0][2].plot(list(range(n)),cycle_A)
axs[0][3].plot(list(range(n)),idt_A)
axs[1][0].plot(list(range(n)),D_B)
axs[1][1].plot(list(range(n)),G_B)
axs[1][2].plot(list(range(n)),cycle_B)
axs[1][3].plot(list(range(n)),idt_B)
axs[0][0].set_title('D_A')
axs[0][1].set_title('G_A')
axs[0][2].set_title('cycle_A')
axs[0][3].set_title('idt_A')
axs[1][0].set_title('D_B')
axs[1][1].set_title('G_B')
axs[1][2].set_title('cycle_B')
axs[1][3].set_title('idt_B')
plt.tight_layout()
plt.legend()
fig.savefig('loss_plt_seperate.png')
```
